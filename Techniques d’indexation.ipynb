```python

!pip install PyPDF2
import PyPDF2

def extract_text_from_pdf(pdf_file):
    with open(pdf_file, 'rb') as file:  # Ouverture du fichier PDF en mode binaire
        reader = PyPDF2.PdfReader(file)
        text = ""  # Initialisation d'une variable pour stocker le texte extrait
        for page in range(len(reader.pages)):  # Boucle à travers toutes les pages du PDF
            text += reader.pages[page].extract_text()  # Extraction du texte de chaque page
        return text
pdf_file = "coursentrepreneuriat2024(1).pdf" # Nom du fichier PDF à traiter
texte_pdf = extract_text_from_pdf(pdf_file) # Appel de la fonction pour extraire le texte
print(texte_pdf) # Affichage du texte extrait


```

```python
import re

def nettoyer_texte(texte):
    # Remplace plusieurs espaces consécutifs par un seul espace
    texte = re.sub(r'\s+', ' ', texte)
    # Supprime les numéros de page (par exemple "1 Page 1")
    texte = re.sub(r'\d+\s+Page\s+\d+', '', texte)
    return texte.strip() # Retourne le texte nettoyé

# The call to the functions should be outside the function definition
# Assuming extract_text_from_pdf and pdf_file are defined elsewhere
texte_pdf = extract_text_from_pdf(pdf_file)
texte_nettoye = nettoyer_texte(texte_pdf) # Nettoie le texte extrait
print(texte_nettoye) # Affiche le texte nettoyé
```

```python
import xml.etree.ElementTree as ET

def convertir_en_xml(titre, auteur, contenu):
    racine = ET.Element("document")
    ET.SubElement(racine, "titre").text = titre
    ET.SubElement(racine, "auteur").text = auteur
    ET.SubElement(racine, "contenu").text = contenu

    arbre = ET.ElementTree(racine)

    arbre.write("document.xml", encoding="utf-8", xml_declaration=True)
convertir_en_xml("Titre Exemple", "Auteur Inconnu", texte_nettoye)
```

```python
def creer_index(contenu):
    mots = contenu.split()
    index = {}
    for pos, mot in enumerate(mots):
        mot_lower = mot.lower()
        if mot_lower not in index:
            index[mot_lower] = []
        index[mot_lower].append(pos)
    return index
contenu = " ".join(ET.parse('document.xml').findtext('contenu').split())
index = creer_index(contenu)
print(index)
```

```python
def rechercher_dans_index(mot, index):

    return index.get(mot.lower(), [])

mot_a_rechercher = "isolateur"
positions = rechercher_dans_index(mot_a_rechercher, index)

print(f"Le mot '{mot_a_rechercher}' apparaît aux positions {positions}")
```
